# Kalman Filter

This code implements a 1-D Kalman filter modelling Brownian motion.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
actual_t = [float(i)/5.0 for i in range(101)]
actual_x = [math.sin(x) for x in actual_t ]

sample_t = [float(i) for i in range(21)]
sample_x = [math.sin(s) + 0.5 * np.random.randn() for s in sample_t]

plt.plot(actual_t, actual_x)
plt.plot(sample_t, sample_x, 'o')
```

```python
kalman_x = []
kalman_p = []

p = 0.5
x = sample_x[0]

kalman_x.append(x)
kalman_p.append(p)

q = 0.5
r = 0.5

for i in range(1,len(sample_t)):
    
    x_pred = x
    p_pred = p + q
    
    v = sample_x[i] - x_pred
    s = r + p_pred
    
    k = p_pred / s
    x = x_pred + k * v
    p = p_pred * (1 - k)
    
    kalman_x.append(x)
    kalman_p.append(p)
```

```python
plt.plot(actual_t, actual_x)
plt.plot(sample_t, sample_x, 'o')
plt.plot(sample_t, kalman_x, '-ro')
```
